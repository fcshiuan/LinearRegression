```python
import pandas as pd

df = pd.read_csv('學生身高平均值(6歲-15歲).csv')
df.head(3)
```

```python
df=df.drop(['學年度',"總計","男"],axis=1)
df.head(3)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd

age = df.drop(['女'],axis=1)
h = df['女']

lm = LinearRegression()
lm.fit(age,h)
print('迴歸係數:',lm.coef_)
print('截距:',lm.intercept_)
```

```python
preH = lm.predict(age)

newAge = pd.DataFrame( np.array([11,8,7]) ,columns=['年齡'])
newAgePreH = lm.predict(newAge)
print("AI預測答案:",newAgePreH)
```

```python
import matplotlib.pyplot as plt
plt.scatter(age['年齡'],h)
plt.plot(age['年齡'],preH,color='blue')
plt.plot(newAge,newAgePreH,color='red',marker='o',markersize=5)
```
